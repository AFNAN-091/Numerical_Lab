```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def func(x,y):
    n = np.size(x)
    m_x = np.mean(x)
    m_y = np.mean(y)
    
    ss_xy = np.sum(y*x) - n*m_y*m_x
    ss_xx = np.sum(x*x) - n*m_x*m_x
    
    b_1 = ss_xy/ss_xx
    
    b_0 = m_y - b_1*m_x
    
    return(b_0,b_1)

    

def regression_line(x,y,b):
    plt.scatter(x,y,color = "blue",marker= "o",s = 30)
    
    y_pre = b[0] + b[1]*x
    
    plt.plot(x,y_pre,color = "g")
    
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
x = np.array([0.698132,0.959931,1.134464,1.570796,1.919862])
y = np.array([0.188224,0.209138,0.230052,0.250965,0.313707])

b = func(x,y)
print(b[0])
print(b[1])
regression_line(x,y,b)

```
